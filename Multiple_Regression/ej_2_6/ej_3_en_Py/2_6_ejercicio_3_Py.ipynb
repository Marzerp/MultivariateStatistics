```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
datos = pd.read_csv('Precios_de_viviendas.csv')
datos.head()
```

```python
X = sm.add_constant(datos[['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9']])
print('X:\n', X[0:4])
y = datos['y']
print('y:\n', y[0:4])
M1 = sm.OLS(y, X).fit()
print('y = ', M1.params['const'], '+', M1.params['x1'], '*x1','+', M1.params['x2'], '*x2',
         '+', M1.params['x3'],'*x3','+',M1.params['x4'],'*x4','+', M1.params['x5'], '*x5',
         '+', M1.params['x6'],'*x6','+',M1.params['x7'],'*x7','+', M1.params['x8'], '*x8',
         '+', M1.params['x9'], '*x9')
# RESPUESTA
#y =  14.927647593124268 + 1.9247215624532625 *x1 + 7.000534203316187 *x2 + 0.14917792844699473 *x3 
#     + 2.722807895783589 *x4 + 2.006684019383746 *x5 + -0.41012375543362367 *x6 + 
#     -1.4032352958394725 *x7 + -0.03714908044311099 *x8 + 1.5594466324195757 *x9
datos['y_pred'] = M1.predict(X)
datos
```

G) Trazar una gráfica de probabilidad normal de los residuales. 
 ¿Parece haber algún problema con la hipótesis de normalidad?

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.gofplots import ProbPlot

# Gráfico de residuos vs valores ajustados (equivalente a which=1 en R)
def plot_residuals_vs_fitted(model):
   # Valores ajustados y residuos
    fitted_values = model.fittedvalues
    residuals = model.resid    
    # Crear el gráfico
    plt.figure(figsize=(10, 6))
    plt.scatter(fitted_values, residuals, alpha=0.7)
    plt.axhline(y=0, color='red', linestyle='--')
    plt.xlabel('Valores Ajustados')
    plt.ylabel('Residuos')
    plt.title('Residuos vs Valores Ajustados')
    plt.grid(True, alpha=0.3)
    # Añadir línea de tendencia suavizada (loess)
    sns.regplot(x=fitted_values, y=residuals, scatter=False, lowess=True, 
                line_kws={'color': 'green', 'alpha': 0.8})
    plt.savefig('grafica_Residuos_vs_valores_ajustados.jpg', dpi=300, bbox_inches='tight')
    plt.show()

plot_residuals_vs_fitted(M1)
# No parece haber problemas 
```

```python
from scipy import stats

# Test de Shapiro-Wilk (equivalente a shapiro.test())
def shapiro_test_residuals(model):
    residuals = model.resid    
    # Realizar test de Shapiro-Wilk
    stat, p_value = stats.shapiro(residuals)
    print("Test de Normalidad de Shapiro-Wilk:")
    print(f"Estadístico W = {stat:.6f}")
    print(f"Valor p = {p_value:.6f}")
    # Interpretación
    alpha = 0.05
    if p_value > alpha:
        print("No se rechaza H0: Los residuos siguen una distribución normal")
    else:
        print("Se rechaza H0: Los residuos NO siguen una distribución normal")
    return stat, p_value

shapiro_stat, shapiro_p = shapiro_test_residuals(M1)
#Test de Normalidad de Shapiro-Wilk:
#Estadístico W = 0.962319
#Valor p = 0.486753
#No se rechaza H0: Los residuos siguen una distribución normal
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt

def grafica(variable, nombre_archivo=None):
    """
    Genera un gráfico de residuales vs una variable especificada
    
    Parameters:
    variable (str): Nombre de la variable a graficar en el eje x
    nombre_archivo (str): Nombre del archivo para guardar la gráfica (opcional)
    """
    plt.figure(figsize=(10, 6))
    plt.scatter(datos[variable], M1.resid, marker='o', color='blue', alpha=0.7)
    plt.title(f'Residuales vs {variable}', fontsize=14, fontweight='bold')
    plt.xlabel(f'Variable {variable}', fontsize=12)
    plt.ylabel('Residuales', fontsize=12)
    plt.axhline(y=0, color='red', linestyle='--', linewidth=1)  # línea en y=0
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    
    # Guardar la gráfica si se especifica un nombre de archivo
    if nombre_archivo:
        plt.savefig(nombre_archivo, dpi=300, bbox_inches='tight')
        print(f"Gráfica guardada como: {nombre_archivo}")
    
    plt.show()

grafica('x1', 'archivo_residuales_vs_x1.jpg')
grafica('x2', 'archivo_residuales_vs_x2.jpg')
grafica('x3', 'archivo_residuales_vs_x3.jpg')
grafica('x4', 'archivo_residuales_vs_x4.jpg')
grafica('x5', 'archivo_residuales_vs_x5.jpg')
grafica('x6', 'archivo_residuales_vs_x6.jpg')
grafica('x7', 'archivo_residuales_vs_x7.jpg')
grafica('x8', 'archivo_residuales_vs_x8.jpg')
grafica('x9', 'archivo_residuales_vs_x9.jpg')

# Parece que la variable x7 se podría omitir, puesto que la x6 aporta información relacionada. 
```
